List of Depths: Given a binary tree, design an algorithm which creates a linked list of all the nodes
at each depth (e.g., if you have a tree with depth 0, you'll have 0 linked lists).

```python
class TreeNode():
    def __init__(self,root=None,left=None,right=None):
        self.root = root
        self.left = left
        self.right = right
    
class Node():
    def __init__(self,root,next=None):
        self.val = root
        self.next = next

def build(arr):
    if arr == []:
        return 
    curr = Node(arr[0])
    start = curr
    i = 1
    while i < len(arr):
        tmp = Node(arr[i])
        curr.next = tmp
        curr = curr.next
        i+=1
    return start

def ListofDepths(head):
    list_of_depths = [Node(head.root),]
    queue = [head]
    tmpqueue = []
    while queue:
        curr = queue.pop(0)
        if curr.left:
            tmpqueue.append(curr.left)
        if curr.right:
            tmpqueue.append(curr.right)
        if queue == []:
            list_of_depths.append(build([x.root for x in tmpqueue]))
            print(tmpqueue)
            queue.extend(tmpqueue)
            tmpqueue = []
    return list_of_depths[:-1]
    
    
```

```python
tree = TreeNode(1,TreeNode(2,TreeNode()),TreeNode(3))
```

```python
x = ListofDepths(tree)
```
