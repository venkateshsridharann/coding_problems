```python

def Anagram_sort(anagram_list):
    out = []
    dict = {}
    for word in anagram_list:
#          O(N*M)
        key = 0
        for i in word:
            key += ord(i)
        if key in dict:
            dict[key].append(word)
        else:
            dict[key] = [word,]
    for key in dict:
#         O(N)
        out.extend(dict[key])
    return dict
```

```python
analist = ['cat','dog','god','tac','act','pod','trick','dop']
Anagram_sort(analist)
```

```python
def partitions(s):
    if len(s) > 0:
        for i in range(1, len(s)+1):
            first, rest = s[:i], s[i:]
            for p in partitions(rest):
                yield [first] + p
    else:
        yield []
```

```python
x = 'aba'
def codify(x):
    
    unique_chars = list(set(x))
    encoding = {}
    i = 1
   
    for c in unique_chars:
        encoding[c] = i
        i = i+1
   
    result = ''
    for c in x:
        result += str(encoding[c])
        
    return result
```
